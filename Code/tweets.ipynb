```python
presentday = datetime.date.today()
pastday = presentday - datetime.timedelta(days = 1)
presentday, pastday
```

```python
tweetslist = tweepy.Cursor(api.search_tweets, q = "#covid-19 since:" + str(pastday) + "until:" + str(presentday), tweet_mode = 'extended', lang = 'en').items()
```

```python
output = []
for tweet in tweetslist:
    text = tweet._json["full_text"]
    print(text)
    favourite_count = tweet.favorite_count
    retweet_count = tweet.retweet_count
    created_at = tweet.created_at
    
    line = {'text' : text, 'favourite_count' : favourite_count, 'retweet_count' : retweet_count, 'created_at' : created_at}
    output.append(line)
```

```python
import pandas as pd
df = pd.DataFrame(output)
df.to_csv('output.csv', mode='a', header=False)
```
